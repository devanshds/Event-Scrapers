```python
import requests
import re
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get("https://www.eventshigh.com/")
c=r.content
```

```python
soup=bs(c,"html.parser")
print(soup.prettify())
```

```python
all=soup.find_all("div",{"class":"d-inline-block event-card-wrp valign-top ga-card-track"})
print(all)
```

```python
links = []
for div in all:
    link=div.find('a')
    links.append('https://www.eventshigh.com/'+link.get('href'))
print(links)
```
