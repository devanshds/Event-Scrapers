```python
import pandas as pd
import requests
import re
import json
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get("https://www.eventshigh.com/")
c=r.content
```

```python
soup=bs(c,"html.parser")
```

```python
int_url = []
non_int_url = []
for link in soup.find_all('a'):
    str1=str(link.get('href'))
    if str1.startswith('/detail/') and len(int_url)<10:
        int_url.append("https://www.eventshigh.com"+link.get('href'))
    elif len(int_url)==10:
        break
    else:
        non_int_url.append("https://www.eventshigh.com"+link.get('href'))
df=pd.DataFrame()
df["Non Interesting URLs"]=pd.Series(non_int_url)
df["Interesting URLs"]=pd.Series(int_url)
df
```

```python
""" Name, By, Description, Daybeg, Dayend, Location, Price """
l= []
for url in int_url:
    d= {}
    r1=requests.get(url)
    c1=r1.content
    page_soup = bs(c1,'html.parser')
    data = page_soup.select("[type='application/ld+json']")[2]
    oJson = json.loads(data.text)
    numProducts = len(oJson)
    product=oJson[0]
    d["Name"]=product['name']
    d["Description"]=product['description']
    d["Start Date"]=product['startDate']
    d["End Date"]=product['endDate']
    d["Location"]=product['location']['name']+','+product['location']['address']
    l.append(d)
df=pd.DataFrame(l)
df
```
